```python
import ee
import geemap
```

```python
Map = geemap.Map()
roi = geemap.shp_to_ee(r"D:\物候分析\侵蚀区提取\研究区数据\DBATP\DBATP.shp").geometry()
```

```python
years = [i for i in range(2001, 2025)]
precipitation = ee.ImageCollection("NASA/GPM_L3/IMERG_MONTHLY_V07") \
            .filterBounds(roi) \
            .filterDate('2001-01-01', '2025-01-01') \
            .select('precipitation')

soil_temperature = ee.ImageCollection("ECMWF/ERA5_LAND/MONTHLY_AGGR") \
            .filterBounds(roi) \
            .filterDate('2001-01-01', '2025-01-01') \
            .select('soil_temperature_level_1') \
            .map(lambda img : img.divide(273.15).copyProperties(img, ['system:time_start']))

datas = [precipitation, soil_temperature]
```

```python
maskimg = ee.Image('projects/ee-qingqiu760/assets/qingz').reproject(crs='EPSG:4326', scale=1000).mask()
```

```python
for year in years:
    dataset = precipitation.filter(ee.Filter.calendarRange(year, field = 'year')).sum() \
                    .reproject(crs='EPSG:4326', scale=1000).updateMask(maskimg).clip(roi)
    geemap.download_ee_image(dataset, f"D:\物候分析\气象湿度\栅格数据\pre_{year}.tif", roi, 'EPSG:4326', scale=1000)
```
