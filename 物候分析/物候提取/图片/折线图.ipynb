```python
# 绘图
import numpy as np
import pandas as pd
from scipy import stats
from scipy.stats import linregress
from matplotlib import pyplot as plt
from matplotlib import font_manager
```

```python
chinese_font = font_manager.FontProperties(fname=r'C:/Windows/Fonts/simsun.ttc')  # 替换路径为实际宋体路径
plt.rcParams['font.family'] = ['Times New Roman']
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题
plt.rcParams['font.size'] = 14    # 设置全局字体大小
```

```python
dt = pd.read_csv(r"D:\物候分析\物候提取\均值.csv")
dt = dt.set_index('Unnamed: 0')
```

```python
x = dt.index
y = dt['log']
slope,intercept,rvalue,pvalue,stderr=stats.linregress(x,y)
def myfunc(x):
    return slope*x+intercept
getmodel=list(map(myfunc,x))
```

```python
plt.figure(figsize=(8,5), num='(a)')
plt.plot(x, y, color='black', linewidth=2, marker='8')
plt.xticks(x, x, rotation=45)
txt = f'y = {round(slope, 4)}x - {round(-intercept, 2)}'
plt.text(2001, 123, txt)
plt.text(2001, 121, f'R² = {round(rvalue, 4)}')
plt.plot(x,getmodel, color='black')
plt.xlabel('年份（year）', fontproperties=chinese_font)
plt.ylabel('生长季长度LOG', fontproperties=chinese_font)
plt.show()
```
