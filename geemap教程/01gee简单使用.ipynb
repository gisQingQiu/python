```python
import ee
import geemap
```

# 授权以及初始化Earth Engine

```python
# ee.Authenticate()    # 授权
ee.Initialize()    # 初始化
```

# geemap 中 Map 对象的简单使用

```python
Map = geemap.Map()    # 创建地图对象
Map.setCenter(116.391, 39.906, 4)    # 设置地图中心和缩放级别
Map    # 显示地图
```

## 查阅官方文档查看 Map 对象的方法
### https://developers.google.com/earth-engine/apidocs

```python
Map.add_basemap("SATELLITE")    # 添加底图
print(Map.get_center())    # 获取地图中心
print(Map.get_zoom())    # 获取地图缩放级别
print(Map.layers)    # 获取图层列表
```

```python
Map.set_zoom(6)    # 设置地图缩放级别
Map.remove(Map.layers[-1])    # 移除最后一个图层
Map
```

# geemap 底图

```python
basemaps = geemap.basemaps    # 获取所有底图
for basemap in basemaps:
    print(basemap)    # 打印每个底图的名称
```

```python
Map.add_basemap("Esri.WorldStreetMap")    # 添加 Esri.WorldStreetMap 底图
# 也可以在交互式地图侧边工具栏中选择底图
Map
```

## 交互式地图侧边工具栏中工具应用（部分）
**1. Inspector 识别工具，用于获取图层中指定位置的属性值**  
2. Change basemap 快速更改底图  
**3. Convert Earth Engine JavaScript to Python 将 JS 脚本转换为 Python 脚本**   
4. Open local vector / raster data 打开本地的数据  
5. Creating and plotting transects 绘制线段后查看指定影像的指定波段，绘制像元值折线图

```python
Map.addLayer(ee.Image("USGS/SRTMGL1_003"), {}, "SRTM DEM")
```

# 绘图工具的使用
### 交互式地图左侧栏中有绘图工具，可以在地图上绘制范围，用于创造要素

```python
Map.remove('SRTM DEM')    # 移除 SRTM DEM 图层
Map
```

```python
ee.FeatureCollection(Map.draw_features)    # 获取绘制的要素集
```

```python
roi = Map.draw_last_feature    # 获取最后绘制的要素
```

```python
Map.addLayer(roi, {}, "ROI")    # 添加 ROI 图层
```

```python
Map.addLayer(roi, {'color': 'blue'}, "ROI")    # 将 ROI 添加到地图
```
